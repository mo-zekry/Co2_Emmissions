```python
df=df[['TIME','Value']]
df.head()
```

```python
df.plot(style='.',
        figsize=(15, 5),
        color=color_pal[0],
        title='Co2 emissions')
plt.show()
```

```python
df[['value of Co2 emissions']].plot(figsize=(20,15),marker='o')

plt.title('Carbon dioxide emissions from cars in World',fontsize=30)
plt.xlabel('change over Time',fontsize=30)
plt.ylabel('co2 Emissions',fontsize=30)
plt.legend()
plt.show()
```
